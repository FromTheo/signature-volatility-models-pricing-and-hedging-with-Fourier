## Asian options

```python
y, kappa, theta, eta, alpha, rho = 0.25, 1, 0.25, 1.2, 0.6, -0.6
M = 4
bell = compute_l_mGBM(y, kappa, theta, eta, alpha, M) 

N = 2500
n = 252
S0 = 1
M = 4
L = 32
```

### 1. $T = 1$ month

```python
T = 1/12
W, paths = generate_paths(N, n, T)
W_perp = generate_brownian_paths(N, n, T)
B = rho * W + np.sqrt(1 - rho**2) * W_perp
```

#### 1.1 $K = 1$

```python
K = 1 
```

```python
PnL_Sig, PnL_BS = quadratic_hedging_PnL_asian(y, bell, rho, S0, K, T, L, M, n, W, B, paths, sigma_BS=0.25, flag="put")
```

```python
np.mean(PnL_Sig**2), np.mean(PnL_BS**2)
```

```python
plot_PnL_distribution(PnL_BS, PnL_Sig, title=r"$T = 1 \text{ month},\; K = 1$")
```
